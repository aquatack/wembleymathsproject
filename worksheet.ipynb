# Wembley Stadium Project 
## Seat Sales
We need to work out what price to sell tickets for each of the four bands available in the stadium.
First we create special variables (called dictionaries in this case), to represent the choices.

```python
# The seats available for each band are stored in this first dictionary called SeatsAvailable. The keys (the first number) represent the band numbers, while the values (the second number) are the number of seats available for that band. 
SeatsAvailable = {
    1:20000,
    2:30000,
    3:10000,
    4:20000
}  

# The BandPriceOptions dictionary contains the price options for each band, where the keys are the band numbers and the values are dictionaries containing the price options and their corresponding seat sales percentages.
BandPriceOptions = {
    1: {50: 100, 65:85, 90:60, 100:40, 110:20, 120:10},
    2: {40: 100, 50:85, 75:60, 80:40, 90:20, 100:10},
    3: {25: 100, 40:85, 60:60, 70:40, 85:20, 100:10},
    4: {10: 100, 15:85, 35:60, 45:40, 50:20, 60:10}
}
```

### Sold Seats Calculation
We need to work out how many seats sell depending on the price of the ticket for each band.
20% of those sold seats will be concessions (at a 50% discount), so we need to know how many full price and concession seats we sell given a band and price.

```python

# The CalculateSeatSalesQty function will return the number of seats sold for a given band and price. It returns both the full price and the concession quantities.
# The function takes in four parameters: price, band, seatsAvailable, and bandPriceOptions. 
# It calculates the number of available seats for a given band based on the price and the number of available seats. 
# It also checks if the provided band and price are valid options in the BandPriceOptions dictionary. If they are not valid, it raises a KeyError with an appropriate message.
def CalculateSeatSalesQty(price, band, seatsAvailable, bandPriceOptions):
    if band not in seatsAvailable:
        raise KeyError(f"Unknown band: {band}")
    if band not in bandPriceOptions or price not in bandPriceOptions[band]:
        raise KeyError(f"Unknown price option for band {band}: {price}")
    
    bandAvailableSeatQty = seatsAvailable[band]  # Get the total available seats for the given band
    seatSalesPerc = bandPriceOptions[band][price]  # Get the percentage of tickets sold for the given price   
    soldSeatQty = bandAvailableSeatQty * (seatSalesPerc / 100)  # Adjust available seats based on sales percentage
    concessionSoldSeatQty = soldSeatQty * 0.2  # Calculate the quantity of concession sales (20% of available seats)
    fullPriceSoldSeatQty = soldSeatQty * 0.8  # Calculate the quantity of full price sales (80% of available seats)
    return {"concessionSoldSeatQty": concessionSoldSeatQty, "fullPriceSoldSeatQty": fullPriceSoldSeatQty}  # Return the adjusted number of available seats
```

## Ticket Sales Income
Once we have the quantity of full price and concession seats for a given band, we can work out the income of from full price and concession ticket sales.

```python
# Now we will define a function called CalculateBandSales that will return the total sales revenue for a given band and price. 
# It will use the CalculateSeatSalesQty function to get the sold seat quantities and then calculate the total sales revenue based on the full price and concession sales.
def CalculateBandSales(price, band, seatsAvailable, bandPriceOptions):
    if band not in seatsAvailable:
        raise KeyError(f"Unknown band: {band}")
    if band not in bandPriceOptions or price not in bandPriceOptions[band]:
        raise KeyError(f"Unknown price option for band {band}: {price}")
    
    soldSeatQty = CalculateSeatSalesQty(price, band, seatsAvailable, bandPriceOptions)  # Get the sold seat quantities for the given band and price
    concessionSoldSeatQty = soldSeatQty["concessionSoldSeatQty"]  # Get the quantity of concession sales
    fullPriceSoldSeatQty = soldSeatQty["fullPriceSoldSeatQty"]  # Get the quantity of full price sales
    
    fullPriceSales = fullPriceSoldSeatQty * price  # Calculate full price sales revenue
    concessionSales = concessionSoldSeatQty * (price * 0.5)  # Calculate concession sales revenue (50% of price)
    totalSales = fullPriceSales + concessionSales  # Calculate total sales revenue
    return totalSales  # Return the total sales revenue
    
```

## Programme Sales
Cost: £7.50
25% of adults (full price sales) buy them
12.5% of concessions buy them

BUT:
for every 200 programmes sold, you need 1 programme seller who works between midday and 3pm. They need to be paid £15 per hour

```python
import math

# Let's create a function which, given the number of full price and concession tickets sold, it will calculate the programme sales revenue.

def CalculateProgrammeSales(fullPriceSoldSeatQty, concessionSoldSeatQty):
    # 25% of the adults buy a programme
    adultProgrammeSalesQty = fullPriceSoldSeatQty * 0.25
    # 12.5% of the concessions buy a programme
    concessionProgrammeSalesQty = concessionSoldSeatQty * 0.125    
    # Calculate total programme sales Qty
    totalProgrammeSalesQty = (adultProgrammeSalesQty + concessionProgrammeSalesQty)
    
    # one programme seller can sell 200 programmes. How many sellers are we going to need?
    programmeSellersNeeded = math.ceil(totalProgrammeSalesQty / 200) # Note that we can't have a fractional person. We'll need to round up to the nearest whole number which is what math.ceil() does (ceil short for ceiling!).
    
    # Programme sellers cost £15 per hour. They work for 3 hours. How much will we pay them in total?
    programmeSellerCost = programmeSellersNeeded * 15 * 3
    
    # Calculate total programme sales revenue
    totalProgrammeSales = (totalProgrammeSalesQty * 7.5) - programmeSellerCost
    
    return totalProgrammeSales
```

### Programme Sales Test

```python
# Let's say we sold 1000 full price tickets and 1000 concession tickets. How much would we make from programme sales?

# 1000 full price tickets sold means 250 programme sales (25% of 1000)
# 1000 concession tickets sold means 125 programme sales (12.5% of 1000)
# Total programme sales = 250 + 125 = 375
# we'll need 2 programme sellers (375 / 200 = 1.875, round up to 2)
# Programme sellers cost £15 per hour. They work for 3 hours. So we will pay them £90 (2 * 15 * 3)
# Total programme sales revenue = 375 * £7.5 - £90 = £2,722.50

programmesales = CalculateProgrammeSales(1000, 1000)
print(f"Total programme sales revenue for 1000 full price and 1000 concession tickets sold: £{programmesales:,.2f}")
```

## Food and Drink
Every Adult will spend on average £7.50 on food and drink. Every concession will spend on average £4 on food and drink

Expenditure for this is:
 - 150 staff working for four hours at £8 per hour
 - £80,000 food and drink bill
 - £2,000 bill for gas, electricity and lighting

```python
# Let's write a function that will calculate the revenue from food and drink

def CalculateFoodAndDrinkSales(fullPriceSoldSeatQty, concessionSoldSeatQty):
    # Every adult spends on average £7.50.
    adultFoodAndDrinkSales = fullPriceSoldSeatQty * 7.5
    # Every concession spends on average £4.00.
    concessionFoodAndDrinkSales = concessionSoldSeatQty * 4.0
    
    # Fixed costs:
    staffSalary = 150*4*8 # 150 staff members, 4 hours, £8 per hour
    foodAndDrinkCost = 80000
    servicesCost = 2000
    
    # Total revenue from food and drink sales
    totalFoodAndDrinkRevenue = adultFoodAndDrinkSales + concessionFoodAndDrinkSales - staffSalary - foodAndDrinkCost - servicesCost
    
    return totalFoodAndDrinkRevenue
```

### Food and Drink Test

```python
# Let's assume 1000 adult and 1000 concession tickets sold. How much would we make from food and drink sales?

# 1000 adult tickets sold means £7,500 from adults (1000 * £7.50)
# 1000 concession tickets sold means £4,000 from concessions (1000 * £4.00)
# Total revenue from food and drink sales = £7,500 + £4,000 - £150*4*8 - £80,000 - £2,000 = -£75,300 (a loss)

foodanddrinkrevenue = CalculateFoodAndDrinkSales(1000, 1000)
print(f"Total revenue from food and drink sales for 1000 adult and 1000 concession tickets sold: £{foodanddrinkrevenue:,.2f}")
```

## Merchandise
Estimated that every adult will spend on average £12. Every concession will spend on average £10.
Expenditure:
 - 75 staff working for 8 hours at £10/hour
 - £315,000 cost of merchandise

```python
# Let's calculate the revenue from Merchandise sales.

def CalculateMerchandiseSales(fullPriceSoldSeatQty, concessionSoldSeatQty):
    # Every adult spends on average £12.00.
    adultMerchandiseSales = fullPriceSoldSeatQty * 12.0
    # Every concession spends on average £10.00.
    concessionMerchandiseSales = concessionSoldSeatQty * 10.0
    
    # Fixed costs:
    staffSalary = 75*8*10 # 75 staff members, 8 hours, £10 per hour
    merchandiseCost = 315000
    
    # Total revenue from merchandise sales
    totalMerchandiseRevenue = adultMerchandiseSales + concessionMerchandiseSales - staffSalary - merchandiseCost
    
    return totalMerchandiseRevenue
```

### Merchandise Revenue Test

```python
# Assume 1000 adult and 1000 concession tickets sold. How much would we make from merchandise sales?
# 1000 adult tickets sold means £12,000 from adults (1000 * £12.00)
# 1000 concession tickets sold means £10,000 from concessions (1000 * £10.00)
# £22,000 income
# Salary costs = 75*8*10 = £6,000
# Merchandise costs = £315,000
# Total costs = £6,000 + £315,000 = £321,000
# Total revenue from merchandise sales = £22,000 - £321,000 = -£299,000 (a loss)

merchandiserevenue = CalculateMerchandiseSales(1000, 1000)
print(f"Total revenue from merchandise sales for 1000 adult and 1000 concession tickets sold: £{merchandiserevenue:,.2f}")
```

## Other Expenditures
 - Gas & Elec = £23,000 + 20% VAT
 - Insurance: £500,000
 - Ticket printing & postage: £2.80 per ticket sold
 - Other staffing: £250,000
 - Contingency: £2million


```python
# Calculate other expenditures

def CalculateOtherExpenditures(fullPriceSoldSeatQty, concessionSoldSeatQty):
    # Fixed costs:
    gasElec = 23000*1.2 # 20% VAT
    Insurance = 500000
    ticketDistribution = 2.80 * (fullPriceSoldSeatQty + concessionSoldSeatQty) # £2.80 per ticket
    staffing = 250000
    contingency = 2000000
    
    # Total other expenditures
    totalOtherExpenditures = gasElec + Insurance + ticketDistribution + staffing + contingency
    
    return totalOtherExpenditures
```

### Other Expenditure Test

```python
# Assume 1000 adult and 1000 concession tickets sold. How much other expenditures would we have?

# Gas and electricity costs = £23,000 + 20% VAT = £27,600
# Insurance costs = £500,000
# Ticket distribution costs = £2.80 * 2000 tickets = £5,600
# Staffing costs = £250,000
# Contingency costs = £2,000,000
# Total other expenditures = £27,600 + £500,000 + £5,600 + £250,000 + £2,000,000 = £2,783,200

otherexpenditures = CalculateOtherExpenditures(1000, 1000)
print(f"Total other expenditures for 1000 adult and 1000 concession tickets sold: £{otherexpenditures:,.2f}")
```

## Bring it altogether
Revenue is:
 - Ticket Sales income
 - Programme Sales Revenue
 - Food and Drink Revenue
 - Merchandise Revenue
 - Other expenditure costs.

```python
# Calculate P&L

# Define the band 1,2,3 and 4 prices
bandPrices = [50,40,25,10]

# check prices are valid
def getAvailablePricingOptions(BandPriceOptions):
    scenarioNumber = 1
    availablePricingOptions = []
    for band1Price in BandPriceOptions[1].keys():
        for band2Price in BandPriceOptions[2].keys():
            if band2Price >= band1Price:
                continue
            for band3Price in BandPriceOptions[3].keys():
                if band3Price >= band2Price:
                    continue
                for band4Price in BandPriceOptions[4].keys():
                    if band4Price >= band3Price:
                        continue
                    availablePricingOptions.append({scenarioNumber: [band1Price, band2Price, band3Price, band4Price]})
                    scenarioNumber += 1
                
    return availablePricingOptions
    


def calculatePL(bandPrices):
    # for each band, get adult and concession ticket sales
    concessionSoldSeatQty  = 0
    fullPriceSoldSeatQty = 0
    ticketSalesIncome = 0

    for (band, price) in enumerate(bandPrices):
        soldSeatQty = CalculateSeatSalesQty(price, band+1, SeatsAvailable, BandPriceOptions)
        concessionSoldSeatQty += soldSeatQty["concessionSoldSeatQty"]
        fullPriceSoldSeatQty += soldSeatQty["fullPriceSoldSeatQty"]
        
        # Calculate ticket sales income
        ticketSalesIncome += CalculateBandSales(price, band+1, SeatsAvailable, BandPriceOptions)
        

    # Calculate programme sales income
    programmeSalesIncome = CalculateProgrammeSales(fullPriceSoldSeatQty, concessionSoldSeatQty)

    # Calculate food and drink sales income
    foodAndDrinkIncome = CalculateFoodAndDrinkSales(fullPriceSoldSeatQty, concessionSoldSeatQty)

    # Calculate merchandise sales income
    merchandiseIncome = CalculateMerchandiseSales(fullPriceSoldSeatQty, concessionSoldSeatQty)

    # Calculate other expenditures
    otherExpenditures = CalculateOtherExpenditures(fullPriceSoldSeatQty, concessionSoldSeatQty)

    # Calculate total revenue and profit/loss
    totalRevenue = ticketSalesIncome + programmeSalesIncome + foodAndDrinkIncome + merchandiseIncome
    profitOrLoss = totalRevenue - otherExpenditures

    # print(f"P&L for Pricing: {bandPrices}")
    # print(f"  Ticket Sales Income: £{ticketSalesIncome:,.2f}")
    # print(f"  Programme Sales Income: £{programmeSalesIncome:,.2f}")
    # print(f"  Food and Drink Income: £{foodAndDrinkIncome:,.2f}")
    # print(f"  Merchandise Income: £{merchandiseIncome:,.2f}")
    # print(f"  Other Expenditures: £{otherExpenditures:,.2f}")
    # print(f"  Profit/Loss: £{profitOrLoss:,.2f}\n")
    return profitOrLoss

# Calculate ticket sales income

# calculate remaining revenue

pricingOptions = getAvailablePricingOptions(BandPriceOptions)
# print(pricingOptions)

#calculate the P&L for each scenario and store the results in a dictionary
plResults = {}
winningScenario = {"scenarioNumber": 1,
                   "scenarioPL": 0}
for scenario in pricingOptions:
    scenarioNumber = list(scenario.keys())[0]
    bandPrices = scenario[scenarioNumber]
    plResults[scenarioNumber] = calculatePL(bandPrices)
    if plResults[scenarioNumber] > winningScenario["scenarioPL"]:
        winningScenario = {"scenarioNumber": scenarioNumber, "scenarioPL": plResults[scenarioNumber], "bandPrices": bandPrices}
        
    

    
print(plResults)
print(f"Winning Scenario: {winningScenario}")
```
